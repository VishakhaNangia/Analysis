```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3,random_state = 101)
print(y.shape)
print(x.shape)
print(y_train.shape)
print(x_train.shape)
print(y_test.shape)
print(x_test.shape)
```

10 Feature Model

```python
x_AverageCC2 = x_train['AverageCC2']
x_PageCheckin =x_train['PageCheckin']
x_PagePopularitylikes = x_train['PagePopularitylikes']
x_PostShareCount = x_train['PostShareCount']
x_PostpublishedSunday = x_train['PostpublishedSunday']
x_BaseDateTimeFriday = x_train['BaseDateTimeFriday']
x_PostpublishedWednesday = x_train['PostpublishedWednesday']
x_Pagetalking =x_train['Pagetalking about']
x_PageCategory = x_train['PageCategory']
x_Basetime = x_train['Basetime']
```

```python
x = np.c_[np.ones(x_AverageCC2.shape[0]),x_AverageCC2, x_PageCheckin,x_PagePopularitylikes,x_PostShareCount,x_PostpublishedSunday,x_BaseDateTimeFriday,x_PostpublishedWednesday,x_Pagetalking,x_PageCategory,x_Basetime]
```

```python
# Parameters required for Gradient Descent
alpha = 0.01   #learning rate
m = y_train.size  #no. of samples
np.random.seed(10)
theta = np.random.rand(11)  #initializing theta with some random values
```

```python
def gradient_descent(x, y, m, theta, alpha):
    cost_list = []   #to record all cost values to this list
    theta_list = []  #to record all theta_0 and theta_1 values to this list 
    prediction_list = []
    run = True
    cost_list.append(1e10)    #we append some large value to the cost list
    i=0
    while run:
        prediction = np.dot(x, theta)   #predicted y values theta_0*x0+theta_1*x1
        prediction_list.append(prediction)
        error = prediction - y
        cost = 1/(2*m) * np.dot(error.T, error)   #  (1/2m)*sum[(error)^2]
        cost_list.append(cost)
        theta = theta - (alpha * (1/m) * np.dot(x.T, error))   # alpha * (1/m) * sum[error*x]
        theta_list.append(theta)
        if cost_list[i]-cost_list[i+1] < 1e-7:   #checking if the change in cost function is less than 10^(-7)
            run = False

        i+=1
    cost_list.pop(0)   # Remove the large number we added in the begining 
    return prediction_list, cost_list, theta_list
```

```python
prediction_list, cost_list, theta_list = gradient_descent(x, y_train, m, theta, alpha)
theta = theta_list[-1]
```

```python
plt.title('Cost Function J', size = 20)
plt.xlabel('No. of iterations', size=15)
plt.ylabel('Cost', size=15)
plt.plot(cost_list)
plt.show()
```

```python
yp = theta[0] +theta[1]*x[:,1] + theta[2]*x[:,2] + theta[3]*x[:,3] + theta[4]*x[:,4] + theta[5]*x[:,5] + theta[6]*x[:,6] + theta[7]*x[:,7]+theta[8]*x[:,8] + theta[9]*x[:,9] 
```

```python
MSE_equ = ((yp-y_train)**2).mean()  #Using yp from equation of hyperplane
MSE_GD = ((prediction_list[-1]-y_train)**2).mean()  #From Gradient Descent


print('Mean Square Error using equation of hyperplane : {}'.format(round(MSE_equ,3)))
print('Mean Square Error from Gradient Descent prediction : {}'.format(round(MSE_GD,3)))
```
